```python
# encoding - typecast karna int- to - other
# # label encoding
# # one hot encoding
from sklearn.preprocessing import LabelEncoder
# isme humara objective data int me convert ho raha hai
l = ['experience_level','employment_type','job_title','salary_currency','employee_residence','company_location','company_size']
le = LabelEncoder()

for i in l:
  df[i] = le.fit_transform(df[i])
```

```python
X = df.drop(columns=['salary_in_usd'])
y = df['salary_in_usd']
```

```python
X = pd.get_dummies(X)
df.shape
```

```python
X
```

```python
# Train - test split
from sklearn.model_selection import train_test_split
# train_test_split()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# def evaluate_model(name,y_train,y_train_pred,y_test,y_test_pred):
  # pass

from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error

def evaluate_model(model,X_train,y_train,X_test,y_test): # model -> Linear , knn, DT , RF.
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)

  print(f"accuracy for your {model} is: ")

  # R2 Score
  print(f"Train R2 score for {model} is {r2_score(y_train,train_pred)}")
  print(f"Teat R2 score for {model} is {r2_score(y_test,test_pred)}")

  # MAE
  print(f"Train MAE for {model} is {mean_absolute_error(y_train,train_pred)}")
  print(f"Testing MAE for {model} is {mean_absolute_error(y_test,test_pred)}")

  # RMSE
  print(f"Train RMSE for {model} is {root_mean_squared_error(y_train,train_pred)}")
  print(f"Testing RMSE for {model} is {root_mean_squared_error(y_test,test_pred)}")
```

```python
from sklearn.linear_model import LinearRegression , Ridge
LinearRegrassion_model = LinearRegression()
```

```python
LinearRegrassion_model.fit(X_train,y_train)
```

```python
evaluate_model(LinearRegrassion_model,X_train,y_train,X_test,y_test)
```

```python
y_pred = LinearRegrassion_model.predict(X_test)

# both - accuracy bad so under fit
```

```python
y_pred_train = LinearRegrassion_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred) # test data accuracy
r2_score(y_train,y_pred_train) # training data accuracy
```

```python
# Hyperperametter tuning

from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python
ridge_param = {
    "alpha":[0.1,1,10,50,100]

}
```

```python
ridge_search = GridSearchCV(ridge,ridge_param,cv=5)
```

```python
ridge_search.fit(X_train,y_train)
```

```python
best_ridge = ridge_search.best_estimator_
```

```python
evaluate_model(best_ridge,X_train,y_train,X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor()
```

```python
knn_param = {
    'n_neighbors': [1,3,5,7,9,11]

}
```

```python
knn_search = GridSearchCV(knn_model,knn_param,cv=5)
```

```python
knn_search.fit(X_train,y_train)
```

```python
best_knn = knn_search.best_estimator_
```

```python
evaluate_model(best_knn,X_train,y_train,X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT_model = DecisionTreeRegressor()
```

```python
DT_param = {
    'max_depth':[3,5,7,10],
    'min_samples_split':[2,5,10]
}
```
